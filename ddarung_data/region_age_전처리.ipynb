```python
import pandas as pd

Location_people_region = '주민등록인구(연령별_동별)_20230523220802.csv'
df_people = pd.read_csv(Location_people_region, encoding='UTF-8')
```

```python
df_people.head()
```

```python
region = df_people['항목'] == '계'

df_region = df_people[region].iloc[:,1:]
df_region.head()
df_region['동별(2)'].values
```

```python
df_region.index = df_region['동별(2)'].values
df_region = df_region.iloc[:,2:]
df_region
```

```python
df_people.iloc[1:2,3:].values[0]
```

```python
# 서울시 총 데이터와 25개의 행정구이므로 행의 길이가 26으로 알맞음.
df_region.columns = df_people.iloc[1:2,3:].values[0]
df_region = df_region.apply(pd.to_numeric)
df_region.head()
```

```python
# 현재 2022년도 4분기와 2023년도 1분기 자료가 섞여 있음. iloc으로 23년도 자료만 빼오기.

df_region = df_region.iloc[:,22:]
df_region
```

```python
# n0~n4세 n5~n9세 합치기!
df_new_index = ['0대','10대','20대','30대','40대','50대','60대','70대','80대','90대']

for i, index in enumerate(df_new_index):
    df_region[index] = df_region.iloc[:,2*i+1] + df_region.iloc[:,2*i+2]
df_region
```

```python
df_region_age = df_region.loc[:,'0대':'90대']
df_region_age['100세 이상'] = df_region.loc[:,'100세 이상']
df_region_age
```

```python
# 자료 전처리 끝.- dropna()는 하면 안됨! 0도 의미있는 값임.

# 이제 진행할 것은 따릉이 데이터에서 연령별로 얼마나 타는 지에 대해서 모집하기
# 연령별로 이용수 구하고, 그 후 연령별 이용률로 변환
# 이용률료 변환한 따릉이 데이터와 행정구별 인구수를 (0~4, 5~9세의 합) 곱해서 따릉이 잠재유저 점수를 측정
# 그 후 이를 정렬해서 나타내기. (그래프)

df_region_age.to_csv('region_age.csv', index=True, header=True, encoding='utf-8-sig')
```
