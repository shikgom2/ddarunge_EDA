```python
import pandas as pd


따릉1 = '서울특별시 공공자전거 이용정보(22년도 전체)_1.csv'
따릉2 = '서울특별시 공공자전거 이용정보(22년도 전체)_2.csv'

df_따릉1 = pd.read_csv(따릉1, encoding='UTF-8',low_memory=False)
df_따릉2 = pd.read_csv(따릉2, encoding='UTF-8', low_memory=False)
```

```python
df_따릉1
```

```python
df_따릉2
```

```python
df_따릉이 = pd.concat([df_따릉1, df_따릉2])
df_따릉이
```

```python
df_따릉이['연령대코드'].unique()

# 현재 존재 ~10대, 10대, 20대, 30대, 40대, 50대, 60대, 70대이상
# 문제 없는 것으로 판명
```

```python
df_연령별_이용수 = df_따릉이.groupby('연령대코드').count()['이용건수']
df_연령별_이용수
```

```python
df_연령별_이용수 = df_연령별_이용수.drop('기타')
df_연령별_이용수
```

```python
df_연령별_이용수.index = ['10대','20대','30대','40대','50대','60대','70대_이상','0대']

# 열린 데이터 광장 측에서 연령대 코드가

# 기타 : AGE_008 ( 비회원, 외국인, 그 이상 등 ) 을 포함하므로, 제외하는 것이 타당함.
```

```python
df_연령별_이용수
```

```python
서울시_연령별 = 'region_age.csv'

df_서울시_연령별 = pd.read_csv(서울시_연령별, encoding='UTF-8')
```

```python
df_서울시_연령별
```

```python
# 70대 이상 컬럼만들기 
df_서울시_연령별['70대_이상'] = df_서울시_연령별['70대'] + df_서울시_연령별['80대'] + df_서울시_연령별['90대'] + df_서울시_연령별['100세 이상']
```

```python
df_서울시_연령별
```

```python
df_서울시_연령별 = df_서울시_연령별.drop(['70대','80대','90대','100세 이상'], axis=1)
df_서울시_연령별
```

```python

df_서울시_연령별.iloc[:,1:]
```

```python
# df_연령별_이용수로 가중치 구하기.
df_연령별_이용수
df_weight = df_연령별_이용수/df_연령별_이용수.sum()
df_weight.sum
```

```python
# 순서 재배치
import matplotlib.pyplot as plt
df_weight = df_weight[['0대','10대','20대','30대','40대','50대','60대','70대_이상']]

plt.rc('font',family='Malgun Gothic')
df_weight.plot()
plt.show()
```

```python
df_weight
```

```python
# df_연령별_이용수를 이용해서 연령별로 곱하기
df_서울시_연령별['User점수'] = df_서울시_연령별.iloc[:,1:].mul(df_weight, axis=1).sum(axis=1)
```

```python
df_서울시_연령별
```

```python
df_서울시_연령별_점수순 = df_서울시_연령별.sort_values('User점수',ascending=False)
```

```python
#따라서 상위 5개는 송파구, 강서구, 관악구, 강남구, 노원구 순이다!
df_서울시_연령별_점수순
```

```python
df_서울시_연령별_점수순.to_csv('region_user_score.csv',index=False,header=True, encoding='cp949')
```
