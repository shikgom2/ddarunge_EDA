```python
#기존 거치대의 인근 100m 안 공공시설 갯수 카운팅 전처리과정
import pandas as pd
from haversine import haversine #pip install haversine
#haversine 함수는 위도, 경도 2개 사이의 거리를 계산해준다.

d1=pd.read_csv('따릉이정보_수정2.csv', encoding='cp949')
df1=pd.DataFrame(d1)

d2=pd.read_csv('서울시 시설물 정보.csv', encoding='cp949')
df2=pd.DataFrame(d2)


bicycle = [0] * len(df1) #빈 리스트에 for문 돌리며 추가하면 index out of range 오류가 떠서 미리 0으로 채워줬다.
public_space = [0] * len(df2)

for i in range(len(df1)):
    bicycle[i] = (df1.iat[i,4], df1.iat[i,5])

for i in range(len(df2)):
    public_space[i] = (df2.iat[i,4], df2.iat[i,5])


public_space_100m = [0] *len(df1)

for i in range(len(df1)):
    count = 0
    for j in range(len(df2)):
        if haversine(bicycle[i], public_space[j], unit = 'm') <= 100: #100m 안에 공공장소 있으면 count에 1 더해준다.
            count = count + 1
    public_space_100m[i] = count


df1['public_place_in_100m'] = public_space_100m

df1.to_csv('따릉이정보_수정2_인근100m_공공장소추가.csv', index=False, encoding='cp949')
```

```python
#기존 거치대의 인근 100m 안 초중고 갯수 카운팅 전처리과정
import pandas as pd
from haversine import haversine #pip install haversine
#haversine 함수는 위도, 경도 2개 사이의 거리를 계산해준다.

d1=pd.read_csv('ddarung_master.csv', encoding='cp949')
df1=pd.DataFrame(d1)

d2=pd.read_csv('서울초중고학교위치데이터_열삭제.csv', encoding='cp949')
df2=pd.DataFrame(d2)


bicycle = [0] * len(df1) #빈 리스트에 for문 돌리며 추가하면 index out of range 오류가 떠서 미리 0으로 채워줬다.
schools = [0] * len(df2)

for i in range(len(df1)):
    bicycle[i] = (df1.iat[i,4], df1.iat[i,5])

for i in range(len(df2)):
    schools[i] = (df2.iat[i,4], df2.iat[i,5])


schools_100m = [0] *len(df1)

for i in range(len(df1)):
    count = 0
    for j in range(len(df2)):
        if haversine(bicycle[i], schools[j], unit = 'm') <= 100: #100m 안에 공공장소 있으면 count에 1 더해준다.
            count = count + 1
    schools_100m[i] = count


df1['schools_in_100m'] = schools_100m

df1.to_csv('ddarung_master.csv', index=False, encoding='cp949')
```

```python
#서울대학목록.csv 전처리과정_1
import pandas as pd

d1=pd.read_csv('서울대학목록.csv', encoding='cp949')
df1=pd.DataFrame(d1)

columns_to_keep = [4, 9, 18]
df1 = df1.iloc[:, columns_to_keep]

df1['new_geo'] = df1.geometry.str.replace("POINT", "")
df1['new_geo_1'] = df1.new_geo.str.replace("(", "")
df1['new_geo_2'] = df1.new_geo_1.str.replace(")", "")


del df1['geometry']
del df1['new_geo']
del df1['new_geo_1']

df1['위도'] = df1.new_geo_2.str.split(' ').str[2]
df1['경도'] = df1.new_geo_2.str.split(' ').str[1]

del df1['new_geo_2']

df1.to_csv('서울대학목록_열삭제.csv', index=False, encoding='cp949')
```

```python
#기존 거치대의 인근 100m 안 대학교 갯수 카운팅 전처리과정
import pandas as pd
from haversine import haversine #pip install haversine
#haversine 함수는 위도, 경도 2개 사이의 거리를 계산해준다.

d1=pd.read_csv('ddarung_master.csv', encoding='cp949')
df1=pd.DataFrame(d1)

d2=pd.read_csv('서울대학목록_열삭제.csv', encoding='cp949')
df2=pd.DataFrame(d2)


bicycle = [0] * len(df1) #빈 리스트에 for문 돌리며 추가하면 index out of range 오류가 떠서 미리 0으로 채워줬다.
unis = [0] * len(df2)

for i in range(len(df1)):
    bicycle[i] = (df1.iat[i,4], df1.iat[i,5])

for i in range(len(df2)):
    unis[i] = (df2.iat[i,2], df2.iat[i,3])


unis_100m = [0] *len(df1)

for i in range(len(df1)):
    count = 0
    for j in range(len(df2)):
        if haversine(bicycle[i], unis[j], unit = 'm') <= 100: #100m 안에 공공장소 있으면 count에 1 더해준다.
            count = count + 1
    unis_100m[i] = count


df1['unis_in_100m'] = unis_100m

df1.to_csv('ddarung_master.csv', index=False, encoding='cp949')
```
