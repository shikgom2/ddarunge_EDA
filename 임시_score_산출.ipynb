```python
import pandas as pd

따릉_master = 'ddarung_master_short.csv'
따릉_population_density = 'points_and_values_divided_by_cnt.csv'
따릉_floor_area_ratio = 'points_and_values_divided_by_cnt_far.csv'

df_따릉_master = pd.read_csv(따릉_master, encoding='cp949',low_memory=False)
df_따릉_인구밀도 = pd.read_csv(따릉_population_density, encoding='cp949', low_memory=False)
df_따릉_용적률 = pd.read_csv(따릉_floor_area_ratio, encoding='cp949', low_memory=False)
```
